```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
df = pd.read_excel('/content/drive/MyDrive/CodingDojo/Week 3/Week51.xlsx')
print(df)
```

```
#Get values from the group and categories
season = df['Season'].values
played = df['Played'].values
won = df['Won'].values
lost = df['Lost'].values
drawn = df['Drawn'].values
#add colors
colors = ['#FF9999', '#00BFFF','#C1FFC1','#CAE1FF','#FFDEAD']
#X-axis information
years = range(len(season))
barWidth = 1
#plot bars
plt.figure(figsize=(10,7))
ax1 = plt.bar(years, won, color=colors[0], edgecolor='white', width=barWidth, label="wins")
ax2 = plt.bar(years, lost, bottom=np.array(won), color=colors[1], edgecolor='white', width=barWidth, label='losses')
ax3 = plt.bar(years, drawn, bottom=np.array(won)+np.array(lost), color=colors[2], edgecolor='white', width=barWidth, label='draws')
plt.legend()
#Label X and Y axis and title
plt.xticks(years, season)
plt.xticks(rotation = 45)
plt.xlabel("Seasons Played")
plt.ylabel("Games Played")
plt.title("Arsenal's Record Over The Last 28 Years")

plt.show()
```
